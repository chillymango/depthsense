```python
import numpy as np
from pathlib import Path
from util import common
import matplotlib.pyplot as plt
```

```python
base_path: Path = Path("data/pseudo/hypersim/0000001")
```

```python
img: np.ndarray = common.normalize(np.load(base_path / "frame.npy"))
z: np.ndarray = common.normalize(np.load(base_path / "depth.npy"))
n: np.ndarray = common.normalize(np.load(base_path / "normal.npy"))

fig, axs = plt.subplots(1, 3, figsize=(9, 3))
axs[0].imshow(img)
axs[1].imshow(z)
axs[2].imshow(n)

for ax in axs:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
